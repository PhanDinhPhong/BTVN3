```python
from google.colab.patches import cv2_imshow
import cv2

image = cv2.imread("/content/bantay.jpg")
cv2_imshow(image)
cv2.waitKey(0)
```

```python
gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
edge = cv2.Canny(gray, 60, 150,apertureSize = 3)
cv2_imshow(edge)
cv2.waitKey(0)
```

```python
edge = cv2.bitwise_not(edge)
cv2.imwrite("/content/bantay.jpg", edge)
palmline = cv2.imread("//content/bantay.jpg")
img = cv2.addWeighted(palmline, 0.4, image, 0.7, 0)
cv2_imshow(img)
```
